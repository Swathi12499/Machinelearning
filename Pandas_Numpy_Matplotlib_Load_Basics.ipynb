### **Matplotlib**

. The axes contain important information about each plot, such as its x- and y-axis labels, title, and more

. subplot() returns a tuple of length two: a figure object as well as the relevant axes object. We will typically pass figsize as a keyword argument.



```python
from matplotlib.pyplot import subplots
```

```python
fig,ax = subplots(figsize=(8,8))
x=rng.standard_normal(100)
y=rng.standard_normal(100)
ax.plot(x,y)
```

```python
fig,ax = subplots(figsize=(8,8))
ax.scatter(x,y,marker='o');

# semicolon prevents text from being printed as above
```

```python
fig,ax = subplots(figsize=(8,8))
ax.scatter(x,y,marker='o')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Scatterplot of x and y');
```

Contour Plot for **3D data**: ax.contour()

Takes 3 args:

*   Vector X : First Dim
*   Vector Y: Second Dim
*   Matrix whose elements correspond to z vals (Third Dim) for each pair of (x,y) coord.






```python
fig, ax = subplots(figsize=(8,8))
x=np.linspace(-np.pi,np.pi,50) #np.linspace(a,b,n) - returns n nums starting at a and ends at b
y=x
f=np.multiply.outer(np.cos(y),1/(1+x**2))
ax.contour(x,y,f);
```

```python
# Adding more levels

fig, ax = subplots(figsize=(8,8))
ax.contour(x, y, f, levels=45);
```

Heatmap

```python
fig, ax = subplots(figsize=(8,8))
ax.imshow(f);
```

Index doesn't have to be unique

```python
Auto_re.loc['ford galaxie 500', ['mpg', 'origin']]
```

```python
# to create df of cols weight, origin for all cars that have year>80

idx_80=Auto_re['year']>80
Auto_re.loc[idx_80,['weight','origin']]
```

Lambda: Anonymous functions

```python
Auto_re.loc[lambda df: df['year']>80,['weight','origin']]

# since lambda is inside Auto_re.loc -> Auto_re will be passed as arg for df
```

```python
# all cars built after 1980 that achieve greater than 30 miles per gallon
Auto_re.loc[lambda df: (df['year']>80) & (df['mpg']>30), ['weight','origin']]
```

```python
#  retrieve all Ford and Datsun cars with displacement less than 300
Auto_re.loc[lambda df: (df['displacement']>300) & (df.index.str.contains('ford'))
| (df.index.str.contains('datsun')), ['weight','origin']]
```

**Summary: For selections**

*   For String and Boolean selections: loc
*   For Integer based selections: iloc
*   Functional queries to filter rows: loc[] with func(lambda in rows arg)





String Formatting

```python
rng = np.random.default_rng(1)
A = rng.standard_normal((127, 5)) # create a matrix of 127*5 from normal dis
# mean 0 and sd 1
M = rng.choice([0, np.nan], p=[0.8,0.2], size=A.shape)
# randomly generate 0(80%) or nan(20%) for a matrix of size A
A += M #(alters some values in A to be nan)
D = pd.DataFrame(A, columns=['food',
'bar',
'pickle',
'snack',
'popcorn']) # create DF
D[:3]
```

```python
# printing percent of missing values in eacgh col

for col in D.columns:
  print('Column {0} has {1:.2%} missing values'.format(col, np.isnan(D[col]).mean()))
```
